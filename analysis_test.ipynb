```python
def sum_tuples(lists):
    result = {}

    # Iterate over each list of tuples
    for lst in lists:
        # Add values from the current list to the result dictionary
        for key, value in lst:
            result[key] = result.get(key, 0) + value

    return result
```

```python
def multiply_dictionary_values(dictionary, multiplier):
    result = {}

    # Multiply the values and update the dictionary
    for key, value in dictionary.items():
        result[key] = value * multiplier

    return result


def devide_dictionary_values(dictionary, devider):
    result = {}

    # Multiply the values and update the dictionary
    for key, value in dictionary.items():
        result[key] = value / devider

    return result
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def rank_profiles(profiles, description, requirement):
    
    
    # Step 1: Preprocessing
    vectorizer = TfidfVectorizer()
    profile_vectors = vectorizer.fit_transform(profiles)
    desc_vector = vectorizer.transform([description])
    req_vector = vectorizer.transform([requirement])

    # Step 2: Calculate Cosine Similarity for each profile
    cosine_sims = cosine_similarity(desc_vector, profile_vectors)*0.3
    req_sims = cosine_similarity(req_vector, profile_vectors)*0.7
    weighted_sims = cosine_sims * req_sims

    # Step 3: Rank the Profiles
    rankings = sorted(enumerate(weighted_sims[0]), key=lambda x: x[1], reverse=True)

    return rankings

# Example usage
profiles = [
    "Profile 1: Python developer with experience in web development.",
    "Profile 2: Full-stack developer with expertise in Python, Django, and React.",
    "Profile 3: Backend developer skilled in Python, Flask, and SQL databases.",
    "Profile 4: Data scientist with Python and machine learning expertise.",
    "Profile 5: Software engineer experienced in Python and C++.",
    "Profile 6: Front-end developer proficient in HTML, CSS, and JavaScript.",
    "Profile 7: Python developer with knowledge of Django and PostgreSQL.",
    "Profile 8: Mobile app developer specializing in Python and Flutter.",
    "Profile 9: DevOps engineer experienced in Python automation and AWS.",
    "Profile 10: Python developer with expertise in data analysis and visualization."
]
description = "Python developer with experience in web development."
requirements = [
    "2+ years of Python development experience",
    "Strong knowledge of Django framework",
    "Experience with front-end technologies like HTML, CSS, and JavaScript",
    
]

main_rankings = []

for requirement in requirements:
    main_rankings.append(rank_profiles(profiles, description, requirement))
    # main_rankings = sum_tuples(rankings,main_rankings)

main_rankings = sum_tuples(main_rankings)


main_rankings = dict(sorted(main_rankings.items(), key=lambda item: item[1],reverse=True))

# print(main_rankings)
for key,value in main_rankings.items():
    print("key: ",profiles[key],"   ","value: ",value)

# for rank, similarity in main_rankings:
#     print(f"Profile: {profiles[rank]} | Similarity: {similarity}")
```
